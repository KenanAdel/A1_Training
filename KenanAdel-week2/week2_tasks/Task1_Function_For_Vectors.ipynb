```python
import numpy as np

def calculate_cosine_similarity(text1, text2):
    words1 = np.array(text1.lower().split())
    words2 = np.array(text2.lower().split())
    all_words = np.array(sorted(list(set(words1.tolist() + words2.tolist()))))

    print(f"Bag of Words: {all_words}")

    v1 = np.isin(all_words, words1).astype(int)
    v2 = np.isin(all_words, words2).astype(int)
    
    print(f"Vector 1: {v1}")
    print(f"Vector 2: {v2}")

    dot_product = np.dot(v1, v2)
    norm1 = np.linalg.norm(v1)
    norm2 = np.linalg.norm(v2)
    
    if norm1 == 0 or norm2 == 0:
        return 0.0
    
    return dot_product / (norm1 * norm2)

t1 = input("Enter First sentence :")
t2 = input("Enter second sentence :")

similarity = calculate_cosine_similarity(t1, t2)
print(f"Similarity {similarity:.2%}")
```
