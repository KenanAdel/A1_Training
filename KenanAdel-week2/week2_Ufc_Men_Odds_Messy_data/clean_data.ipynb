### 4. Rename the columns

```python
df.rename(columns={
    "fighter1" : "fighter_1",
    "fighter2" : "fighter_2",
    "odds1" : "fighter_1_odds",
    "odds2" : "fighter_2_odds",
    "score" : "win_method",
    "time" : "match_duration",
    "date" : "match_date",
    "class" : "weight_class"
} , inplace=True)
```

-------------------

### 5. Replace the names of the column win_method to make it clear

```python
df['win_method'].unique()
```

```python
df['win_method'] = df['win_method'].replace({
    'PTS': 'Decision', 
    'TKO': 'Technical Knockout',
    'SUBM': 'Submission',
    'KO': 'Knockout',
    'canc.': 'Cancelled',       
    'Draw': 'Draw',
    'postp.': 'Postponed',
    'DQ': 'Disqualification'
})
```

--------------------

### 6. clean the duplicated raws

```python
duplicated = df[df.duplicated()]
print(len(duplicated))
```

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

------------------

### 7. Data type conversion

```python
df.head()
```

```python
df.info()
```

```python
df = df.astype({
    "fighter_1" : str,
    "fighter_2" : str,
    "fighter_1_odds" : float,
    "fighter_2_odds" : float,
    "winner" : str,
    "win_method" : 'category',
    "weight_class" : 'category'
})
```

```python
df["match_date"] = pd.to_datetime(df["match_date"])
```

```python
data_duration = df["match_duration"].str.split(":", expand=True)
df["match_duration_minutes"] = data_duration[0].astype(float) + (data_duration[1].astype(float) / 60)
df["match_duration_minutes"] = df["match_duration_minutes"].round(2)
df = df.drop(columns=["match_duration"])
```

```python
df.head()
```

```python
df.info()
```

-----------------------

### 8. Handling missing values

```python
df.isna().sum()
```

```python
df["fighter_1_odds"] = df["fighter_1_odds"].fillna(0)
df["fighter_2_odds"] = df["fighter_2_odds"].fillna(0)
```

## Last thing save the clean data to a new file

```python
df.to_csv("cleaned_ufcmenodds.csv" ,index=False)
```
