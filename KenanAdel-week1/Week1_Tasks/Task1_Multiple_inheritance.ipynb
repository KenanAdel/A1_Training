```python
class Employee:
    def set_info(self, name, id):
        self.name = name
        self.id = id

    def show_employee_data(self):
        return f"Employee: {self.name} | Id: {self.id}"

class Coder:
    def set_skills(self, language):
        self.language = language

    def show_coding_skill(self):
        return f" Programming Languages : {self.language}"

class SoftwareEngineer(Employee, Coder):
    def work(self):
        return "No Thing Just Trying..."

engineer = SoftwareEngineer()

engineer.set_info("Kenan Adel", 1)     
engineer.set_skills("Python")       

print(engineer.show_employee_data()) 
print(engineer.show_coding_skill())   
print(engineer.work())                
```
